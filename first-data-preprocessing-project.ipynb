# **Aggregating**

```python
series = diamondData['price_float']
```

```python
# mean of values in the series data
series.mean()
```

```python
# minimum value in series data
series.min()
```

```python
# maximum value in series data
series.max()
```

```python
# number of values in series data
series.count()
```

```python
# sum of all values in series data
series.sum()
```

```python
# multiplication of all values in series data
display(series.prod())

print('\n\n', 'Non-infinity production:', sep='')
display(pd.Series(range(1, 10)).prod())
```

```python
# median of the values in series data
series.median()
```

```python
# mode of the values in series data
series.mode()
```

```python
# standard deviation of the values in series data
series.std()
```

```python
# variance (i.e. std ** 2) of the values in series data
series.var()
```

**Relationship Plots**

```python
# Was messing with relationship plots
# carat and price depedning on cut; color legend on right

sns.relplot(
    data = df_filtered,
    x = "price_float", y = "carat", col = "cut",
    hue = "color", style = "color",
)
```

**Distribution Plots**

```python
# distribution of price depending on cut type
sns.displot(
    data = df_filtered, 
    x = "price_float", 
    col = "cut"
)
```
